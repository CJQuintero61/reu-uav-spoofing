# data_parser
This file is a helper script to parse the cleaned CSV files that are given to models
to be able to check for some statistics of the cleaned data to see if there are any
obvious columns that classify the data besides the label column

```python
"""imports"""
import numpy as np
import pandas as pd
```

```python
"""load data"""

file = '../data/model1-cleaned.csv'
data = pd.read_csv(file)

# extract the normal and malicious dataframes
df_normal = data[data['label'] == 0]
df_malicious = data[data['label'] == 1]

print(df_normal.shape)
print(df_malicious.shape)
```

```python
"""run some stats"""

# labels are no longer needed since we already separated the dataframes
df_normal.drop(columns=['label'], inplace=True)
df_malicious.drop(columns=['label'], inplace=True)

# both dataframes should have the same columns, so we can loop through both at the same time

for idx, col in enumerate(df_normal.columns):
    normal_values = df_normal[col].values
    malicious_values = df_malicious[col].values

    normal_mean = np.mean(normal_values)
    normal_std = np.std(normal_values)

    malicious_mean = np.mean(malicious_values)
    malicious_std = np.std(malicious_values)

    mean_diff = normal_mean - malicious_mean
    std_diff = normal_std - malicious_std

    # arbitrary threshold to check for significant differences in mean and std
    if abs(std_diff) > 1.0:
        print(f"Feature: {col}")
        print(f"  Normal Mean: {normal_mean:.4f}, Normal Std: {normal_std:.4f}")
        print(f"  Malicious Mean: {malicious_mean:.4f}, Malicious Std: {malicious_std:.4f}")
        print(f"  Mean Difference: {mean_diff:.4f}, Std Difference: {std_diff:.4f}")
        print()
    
```
